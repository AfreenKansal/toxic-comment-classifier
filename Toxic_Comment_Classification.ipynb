```python
import contractions 
import re
from nltk.stem import WordNetLemmatizer 
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords 
stop_words = set(stopwords.words('english')) 
def preprocess(char):
    words = char.lower()
    words_abb = contractions.fix(words)
    words_w = words_abb.strip()
    words_c = re.sub('[^a-zA-Z]',' ',words_w) 
    words_token = word_tokenize(words_c)
    words_clean = [w for w in words_token if w not in stop_words]
    lemmatizer = WordNetLemmatizer()
    words_lemm = [lemmatizer.lemmatize(w) for w in words_clean]    
    return words_lemm
```

```python
for i in range(0,len(train)):
    char = []
    s = ""
    char = preprocess(train['comment_text'][i])  
    for w in char:
        s = s + w + " "
    train['comment_text'][i] = s 
```

```python
train.to_csv("train_clean.csv")
```

In the data pre-processing, various things were done. Firstly,all the words were converted to lowercase. Then he abbreviations were converted into it's full form.
After that,all the extra white spaces, numbers and punctuations are removed from the text.
The text is then tokenized, stop words are removed and then lemmatized.

## Word Embeddings and Neural Networks

```python
#pip install keras
```

```python
from keras.models import Sequential
from keras.layers import Dense,Embedding,Flatten,BatchNormalization,Dropout
from keras.layers.embeddings import Embedding
```

```python
X = train['comment_text']
y = train.drop(['id','comment_text'],axis=1)
```

```python
from tensorflow.python.keras.preprocessing.text import Tokenizer
from tensorflow.python.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer_obj = Tokenizer()
tokenizer_obj.fit_on_texts(X)
```

```python
max_length = max([len(s.split()) for s in X])
vocab_size = len(tokenizer_obj.word_index) + 1
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
X_train_tokens = tokenizer_obj.texts_to_sequences(X_train)
X_test_tokens = tokenizer_obj.texts_to_sequences(X_test)
```

```python
X_train_pad = pad_sequences(X_train_tokens,maxlen=max_length,padding='post')
X_test_pad = pad_sequences(X_test_tokens,maxlen=max_length,padding='post') 
```

```python
# Model for toxic label
model_toxic = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_toxic.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_toxic.summary()) 

num_epochs = 5
history = model_toxic.fit(X_train_pad,y_train['toxic'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)
pred_toxic = (model_toxic.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(data = pred_toxic).to_csv("toxic.csv") 
```

```python
# Model for severe_toxic label

model_s_toxic = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_s_toxic.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_s_toxic.summary()) 

num_epochs = 5
batch_size=256
history = model_s_toxic.fit(X_train_pad,y_train['severe_toxic'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)  
pred_severe_toxic = (model_s_toxic.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(pred_severe_toxic).to_csv("severe_toxic.csv")
```

```python
# Model for obscene label

model_obscene = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_obscene.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_obscene.summary()) 

num_epochs = 5
batch_size=256
history = model_obscene.fit(X_train_pad,y_train['obscene'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)
pred_obscene = (model_obscene.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(data = pred_obscene).to_csv("Obscene.csv")
```

```python
# Model for insult label

model_insult = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_insult.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_insult.summary()) 

num_epochs = 5
batch_size=256
history = model_insult.fit(X_train_pad,y_train['insult'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)
pred_insult = (model_insult.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(data = pred_insult).to_csv("Insult.csv")
```

```python
# Model for threat label

model_threat = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_threat.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_threat.summary()) 

num_epochs = 5
batch_size=256
history = model_threat.fit(X_train_pad,y_train['threat'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)
pred_threat = (model_threat.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(data=pred_threat).to_csv("Threat.csv")
```

```python
# Model for identity_hate label

model_i_hate = Sequential([
  Embedding(vocab_size,50,input_length = max_length),
    Flatten(),
    BatchNormalization(),
    Dropout(0.3),
    Dense(128,activation='relu'),
    Dense(1,activation='sigmoid')
])

model_i_hate.compile(loss='binary_crossentropy',optimizer='adam',metrics=['binary_accuracy'])
print(model_i_hate.summary()) 

num_epochs = 5
batch_size=256
history = model_i_hate.fit(X_train_pad,y_train['identity_hate'],
                   epochs = num_epochs,
                   verbose=2,
                   validation_split=0.2)
pred_identity_hate = (model_i_hate.predict(X_test_pad) >0.5).astype("int32")
pd.DataFrame(data = pred_identity_hate).to_csv("Identity_hate.csv")
```
